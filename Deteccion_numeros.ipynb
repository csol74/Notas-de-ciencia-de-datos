```python
#Guarde en una variable el api del dataset de nmist y cargue en la variable (X_train, y_train), (X_test, y_test) ese dataset
mnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
#Muestreme los 5 primeros elementos de X_train y y_train
print("Primeros 5 elementos de X_train:")
print(X_train[:5])
print("\nPrimeros 5 elementos de y_train:")
print(y_train[:5])
```

```python
#Muestre la forma de un numero del arreglo X_train y de un numero del arreglo y_train
print("Forma de un numero del arreglo X_train:", X_train[0].shape)
print("Forma de un numero del arreglo y_train:", y_train[0].shape)
```

```python
#Crear una red secuencial densa que tenga una capa de entrada de 32 neuronas, una capa oculta de 16 neuronas. La función de activación de las capas ocultas es relu y la capa de salida van a ser 10 neuronas.
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),  # Capa de entrada
    tf.keras.layers.Dense(32, activation='relu'),  # Capa oculta
    tf.keras.layers.Dense(16, activation='relu'),  # Capa oculta
    tf.keras.layers.Dense(10, activation='softmax')  # Capa de salida
])
#Compile el modelo usando el optimizador adam, la función de pérdida sparse_categorical_crossentropy y la métrica accuracy
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
#Entrene el modelo usando el método fit con los datos de entrenamiento, un número de epochs de 5 y un batch_size de 32
model.fit(X_train, y_train, epochs=5, batch_size=32)
#Evalúe el modelo usando el método evaluate con los datos de prueba
test_loss, test_acc = model.evaluate(X_test, y_test)
print('\nTest accuracy:', test_acc)
```

```python
#Guarde el modelo entrenado con el metodo save y tipo de archivo keras
model.save('mnist_model.keras')
```
